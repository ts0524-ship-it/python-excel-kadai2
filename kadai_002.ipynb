```python
import pandas as pd
```

```python
# データフレームの作成
df = pd.DataFrame({'日付':
                   ['2023-05-17','2023-05-18','2023-05-19','2023-05-20','2023-05-21'],
                   '社員名':['山田','佐藤','鈴木','田中','高橋'],
                   '売上': ['100','200','150','300','250'],
                   '部門':['メーカー','代理店','メーカー','商社','代理店']
                   })
```

```python
# 売上全体の平均を求めて新しい列を作成
df['平均売上'] = df['売上'].astype(int).mean()
```

```python
def prioritize(level):
  result = ''
  # df['平均売上']はデータフレーム全体の平均売上を格納しているため、iloc[0]で値を取得
  average_sales_value = df['平均売上'].iloc[0]
  if level >= average_sales_value + 50:
    result = 'A'
  elif level >= average_sales_value:
    result = 'B'
  else:
    result = 'C'
  return result
df['業績ランク'] = df['売上'].astype(int).apply(prioritize)
```

```python
# エクセルファイルへの書き込み
with pd.ExcelWriter('業績.xlsx', engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='sheet1', index=False)
    workbook = writer.book
    sheet = workbook.active
    # 列幅設定をここに移動
    sheet.column_dimensions['A'].width = 15
    sheet.column_dimensions['F'].width = 15
```
